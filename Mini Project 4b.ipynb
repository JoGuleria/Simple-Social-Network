```python
edges = [
    ('Joe', 'Molly'),
    ('Joe', 'Harini'),
    ('Joe', 'Lijuan'),
    ('Molly', 'Joe'),
    ('Molly', 'Renuka'),
    ('Harini', 'Joe'),
    ('Harini', 'Renuka'),
    ('Harini', 'Lijuan'),
    ('Renuka', 'Harini'),
    ('Renuka', 'Joe'),
    ('Lijuan', 'Renuka'),
    ('Lijuan', 'Joe')
]

G = nx.Graph(edges)
DG = nx.DiGraph(edges)
```

```python
adj_list = {
    'Joe':     ['Molly', 'Harini', 'Lijuan'],
    'Molly':   ['Joe', 'Renuka'],
    'Harini':  ['Joe', 'Renuka', 'Lijuan'],
    'Renuka':  ['Harini', 'Joe'],
    'Lijuan':  ['Renuka', 'Joe']
}

G = nx.Graph(adj_list)
DG = nx.DiGraph(adj_list)
```

```python
adj_matrix = np.array([
# Joe  Molly Harini Renuka Lijuan
  [0,    1,     1,     0,     1],   # Joe
  [1,    0,     0,     1,     0],   # Molly
  [1,    0,     0,     1,     1],   # Harini
  [1,    0,     1,     0,     0],   # Renuka
  [1,    0,     0,     1,     0]    # Lijuan
])

G = nx.from_numpy_array(adj_matrix, create_using=nx.DiGraph)
node_names = ['Joe', 'Molly', 'Harini', 'Renuka', 'Lijuan']
G = nx.relabel_nodes(G, dict(enumerate(node_names)))
```

```python
adj_matrix = nx.to_numpy_array(G, dtype = int)
print(adj_matrix)
```

```python
edge_list = list(G.edges())
print(edge_list)
```

```python
plt.figure(figure = (6,6))
nx.draw(G, with_labels = True, node_color = 'lightblue', edge_color = 'gray', node_size = 700, font_size = 12)
plt.title('Simple Soical Network')
plt.show
```
